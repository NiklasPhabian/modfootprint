```python
import pyhdf.SD
import earthpy
import numpy
import earthpy.plot
```

```python
tile_name = 'data/MOD09.A2021314.1845.006.2021316024411.hdf'
mod09 = pyhdf.SD.SD(tile_name)
```

```python
l1 = mod09.select('500m Surface Reflectance Band 1').get()
l2 = mod09.select('500m Surface Reflectance Band 2').get()
l3 = mod09.select('500m Surface Reflectance Band 3').get()
l4 = mod09.select('500m Surface Reflectance Band 4').get()

layer1 = numpy.stack([l1, l2, l3, l4]) 
```

```python
earthpy.plot.plot_rgb(layer1, rgb=[0, 3, 2], stretch=True)
```
